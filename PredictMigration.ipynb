```python
import pandas as pd  #use it to load CSV files, clean data, and handle missing values.
from sklearn.model_selection import train_test_split  #For splitting your dataset
from sklearn import svm  #Great for classification tasks where you need a decision boundary.
import seaborn as sns    #For prettier, more informative visualizations.
import matplotlib.pyplot as plt   #You use this to build charts like line graphs, scatter plots, or bar charts.
from sklearn.metrics import mean_squared_error   #Measures how far off your model’s predictions are from the actual values. The smaller, the better.
import numpy as np   #Essential for computations in ML.
```

```python
data = pd.read_csv(r"C:\Users\ASUS\Downloads\projectML\migration_nz.csv")
data.head(10)
```

```python
data['Measure'].unique()   #let’s see the unique values ​we have in the “Measure” column
```

```python
data['Measure'] = data['Measure'].replace({
    "Arrivals": 0,
    "Departures": 1,
    "Net": 2
})
```

```python
data['Measure'].unique()
```

```python
data['Country'].unique()
```

```python
data['CountryID'] = pd.factorize(data.Country)[0]
data['CitID'] = pd.factorize(data.Citizenship)[0]
```

```python
data['CountryID'].unique()
```

```python
data.isnull().sum()  #we have some missing values, let’s see how many and where exactly they are
```

```python
data.loc[:, "Value"] = data["Value"].fillna(data["Value"].median())   #I will simply fill these missing values with the median values
```

```python
data.isnull().sum()
```

```python
data.drop('Country', axis=1, inplace=True)
data.drop('Citizenship', axis=1, inplace=True)
```

```python
X = data[['CountryID', 'Measure', 'Year', 'CitID']].to_numpy()
Y = data['Value'].to_numpy()
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=9)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=70,max_features = 3,max_depth=5,n_jobs=-1)
rf.fit(X_train ,y_train)
rf.score(X_test, y_test)
#Now, let’s predict migration using our Machine Learning algorithm and visualize the results
```
